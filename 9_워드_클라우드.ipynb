```python
text = open('./alice.txt').read()
# print(text)
```

```python
from PIL import Image # 이미지에서 마스킹 데이터를 출력
from wordcloud import STOPWORDS # 불용어, 워드 클라우드로 작성할 단어에서 제외할 단어를 지정
import numpy as np
```

```python
 alice_mask = np.array(Image.open('./alice_mask.png'))
 alice_mask
```

```python
stopwords = set(STOPWORDS) # 객체 생성
print(type(stopwords))
# stopwords
```

```python
stopwords.add('said')
# stopwords
```

```python
wc = WordCloud(background_color='white', max_words=2000, stopwords=stopwords, mask=alice_mask)
wordcloud = wc.generate(text)
```

```python
plt.figure(figsize=(15,10))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
import random
```

```python
# word : 워드 클라우드로 표시할 단어
# font_size : 워드 클라우드로 표시할 크기
# position : 워드 클라우드로 표시할 단어의 위치
# orientation : 워드 클라우드로 표시할 단어의 방향
# hwargs : 워드 클라우드로 표시할 단어에 적용될 글꼴의 경로
# hsl(색상, 채도, 밝기)
def gray_color_func(word, font_size, position, orientation=None, **hwargs):
  return 'hsl(0, 0%%, %d%%)' % random.randint(60, 100)
  # 0 : 그레이 의미
```

```python
plt.figure(figsize=(15,10))
plt.imshow(wordcloud.recolor(color_func=gray_color_func, random_state=3), interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
from konlpy.corpus import kolaw
from konlpy.tag import Twitter
import nltk # 말뭉치 라이브러리
t = Twitter()
```

```python
ko_law_text = kolaw.open('constitution.txt').read()
ko_law_text
```

```python
tokens_ko = t.nouns(ko_law_text)
tokens_ko
# 한글자짜리는 무엇인지를 확인하기 어려울 정도
```

```python
# 불용어 리스트(제거할 리스트)
stop_words=['제', '월', '일', '조', '수', '때', '그', '이', '바', '및', '안', '정', '관', '위', '장']
tokens_ko = [each_word for each_word in tokens_ko if each_word not in stop_words]
# stop_words에 없는 word만 each_word에 반복해서 저장
```

```python
# Text() : 문서를 편리하게 탐색할 수 있는 다양한 기능을 제공 (nltk라이브러리에 있는 함수)
ko = nltk.Text(tokens_ko)
ko.vocab().most_common(50) # 빈도수가 높은 단어 50개 리스트로 출력
```

```python
data = ko.vocab().most_common(500) # set형태
tmp_data = dict(data) # 딕셔너리로 변환
tmp_data
```
