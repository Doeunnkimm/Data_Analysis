- concat: row나 column 기준으로 단순하게 이어 붙임
- merge: 특정 고유한 키(unique id)값을 기준으로 병합

```python
df = pd.read_csv("C:/Users/82103/Desktop/IT Diary/python_AI/필요 csv/korean-idol.csv")
df2 = pd.read_csv("C:/Users/82103/Desktop/IT Diary/python_AI/필요 csv/korean-idol-2.csv")
```

```python
df_right = df2.drop([1,3,5,7,9], axis=0)
```

```python
# 1,3,5,7,9 행을 제거하고 남은 행들의 인덱스를 reset
df_right = df_right.reset_index(drop=True)
```

```python
df_right
```

```python
# df에 붙일건데 현재 df_right의 인덱스 숫자 및 갯수가 맞지 않은 상태
pd.concat([df, df_right], axis=1)
# 그냥 concat을 하게 되면 '이름'을 보았을 떄 맞지 않는 것을 볼 수 있음
# -> merge를 사용하자
```

```python
# on='이름': 합칠 때 기준, how='left': 왼쪽(df)기준으로 붙이겠다
# on에는 두 DataFrame 모두 있는 키 값만 들어갈 수 있음
pd.merge(df, df_right, on='이름', how='left')
# 이름으로 합쳤기 떄문에 '이름'은 하나로 나타남
```

```python
pd.merge(df, df_right, on='이름', how='right')
# how='right' : df_right 기준으로 붙였기 때문에 왼쪽에서 부족한 건 NaN으로 표시
```

```python
pd.merge(df, df_right, on='이름', how='inner')
# inner 방식은 두 DataFrame에 모두 키 값이 존재하는 경우만 병합
```

```python
pd.merge(df, df_right, on='이름', how='outer')
# outer 방식은 두 DataFrame에 있는 모든 값들을 보여줌
```

```python
# on할 키 이름은 다르지만, 데이터는 같을 때
df_right.columns = ['성함','연봉','가족수']
```

```python
df_right
```

```python
pd.merge(df, df_right, left_on='이름', right_on='성함', how='outer')
# 기준으로 합치긴 하지만, 대신에 이름과 성함 모두 표시
```

# 25. 원 핫 인코딩(One-Hot-Encoding)

- 원 핫 인코딩은 한개의 요소는 True, 나머지 요소는 False로 만들어주는 기법

```python
blood_map = {'A':0,'B':1,'AB':2,'O':3}
```

```python
df
```

```python
df['혈액형_code'] = df['혈액형'].map(blood_map)
```

```python
df.head()
```

```python
df['혈액형_code'].value_counts()
```

- df['혈액형_code']를 머신러닝 알고리즘에 넣어 데이터를 예측하라고 지시한다면 컴퓨터는 값들간의 관계를 스스로 형성하게 됨
- 만약 B형은 1, AB형은 2라는 값을 가지고 있을 때, 컴퓨터는 'B형 + AB형 = O형'라고 잘못된 관계를 맺을 수 있게 됨
- 별도의 column을 형성해주고 1개의 column에는 True, 나머지는 False라고 넣어줌으로 'A,B,O,AB형의 관계는 독립적이다'라는 카테고리로 표현해주는 방식

```python
pd.get_dummies(df['혈액형_code'])
```
