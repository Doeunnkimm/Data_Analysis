### 네이트 실시간 이슈 키워드
* https://www.nate.com/

```python
driver = webdriver.Chrome("D:/AI_Study/chromedriver.exe") # 크롬창을 띄우는 역할
driver.get('https://www.nate.com/')
html = driver.page_source # html 소스를 저장
```

```python
from bs4 import BeautifulSoup
```

```python
soup = BeautifulSoup(html)
issues = soup.findAll('a',{'class':'ik'})
# print(issue)
```

```python
for issue in issues:
    print(issue.text.split('\n')[1], end=' ')
    # print(issue.text.split('\n')[2], end=' ')
    updown = issue.text.split('\n')[2]
    if updown[:2] == '상승':
        print('↑'+ updown[2:])
    elif updown[:2] == '하락':
        print('↓'+ updown[2:])
    elif updown[:2] == '동일':
        print('-')
    elif updown[:2] == 'new':
        print('new')
    else:
        print(updown)
```

### 네이버 웹툰
* https://comic.naver.com/webtoon/weekday

```python
# 댓글 주소 : https://comic.naver.com/comment/comment?titleId=795487&no=19

from selenium import webdriver
from selenium.webdriver.common.keys import Keys
```

```python
driver = webdriver.Chrome("D:/AI_Study/chromedriver.exe")
driver.get('https://comic.naver.com/comment/comment?titleId=795487&no=19')
```

```python
from bs4 import BeautifulSoup
html = driver.page_source # 불러온 페이지의 html 소스를 다 불러옴
soup = BeautifulSoup(html)
```

```python
# 베스트 댓글 파싱
comment_area = soup.findAll('span', {'class': 'u_cbox_contents'})
print(comment_area)
```

```python
print('*****베스트 댓글*****')
for i in range(len(comment_area)):
    comment = comment_area[i].text.strip()
    # 텍스트만 가져와서 앞뒤 공백 없애서 저장
    print(comment)
    print('-'*50)
```

```python
# 전체 댓글 더보기
driver.find_element('xpath', '//*[@id="cbox_module_wai_u_cbox_sort_option_tab2"]/span[2]').click()
```

```python
import time

pagenum = 1

while(True):
    time.sleep(1)
    html = driver.page_source
    soup = BeautifulSoup(html)
    
    comment_area = soup.findAll('span', {'class', 'u_cbox_contents'})
    # print(comment_area)

    for i in range(len(comment_area)):
        comment = comment_area[i].text.strip()
        print(comment)
        print('-'*50)
    
    for i in range(3, 11):
        driver.find_element('xpath', '//*[@id="cbox_module"]/div/div[7]/div/a[' + str(i) + ']').click()
        time.sleep(1)
        html = driver.page_source
        soup = BeautifulSoup(html)
        comment_area = soup.findAll('span', {'class', 'u_cbox_contents'})
        
        for j in range(len(comment_area)):
            comment = comment_area[j].text.strip()
            print(comment)
            print('-'*50)
    if pagenum == 1:
        driver.find_element('xpath','//*[@id="cbox_module"]/div/div[7]/div/a[11]/span[1]').click()
        pagenum += 1
    else:
        driver.find_element('xpath','//*[@id="cbox_module"]/div/div[7]/div/a[12]/span[1]').click()
```
